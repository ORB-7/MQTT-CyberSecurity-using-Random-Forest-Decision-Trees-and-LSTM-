## Implement Proposed Feature Engineering

### Subtask:
Implement key feature engineering steps: creating categorical bins for `tcp_len` and `mqtt_msg`, calculating interaction features like `mqtt_msg_per_tcp_len` and `tcp_len_mqtt_msg_product`, and extracting derived features from `tcp_flags`, `mqtt_conflags`, `mqtt_kalive`, and `mqtt_qos`.


**Reasoning**:
To implement the proposed feature engineering strategies, I will create a copy of the original DataFrame, generate new categorical, interaction, and derived features based on the specified logic, perform timestamp-based feature engineering, remove the constant 'mqtt_dupflag' column, and then display the head and information of the engineered DataFrame along with value counts for the new categorical/binary features to verify the transformations.



````python
import numpy as np

# 1. Create a copy of the original DataFrame `df` and name it `df_engineered`.
df_engineered = df.copy()

# 2. Create a new categorical feature `tcp_len_category`
df_engineered['tcp_len_category'] = pd.cut(df_engineered['tcp_len'],
                                          bins=[0, 54, 600, np.inf],
                                          labels=['small', 'medium', 'large'],
                                          right=True,
                                          include_lowest=True)

# 3. Create a new categorical feature `mqtt_msg_category`
df_engineered['mqtt_msg_category'] = pd.cut(df_engineered['mqtt_msg'],
                                           bins=[-1, 0, 100, 1000, np.inf],
                                           labels=['none', 'low', 'moderate', 'high'],
                                           right=True,
                                           include_lowest=True)

# 4. Calculate the `mqtt_msg_per_tcp_len` feature
df_engineered['mqtt_msg_per_tcp_len'] = df_engineered['mqtt_msg'] / (df_engineered['tcp_len'] + 1)

# 5. Calculate the `tcp_len_mqtt_msg_product` feature
df_engineered['tcp_len_mqtt_msg_product'] = df_engineered['tcp_len'] * df_engineered['mqtt_msg']

# 6. Create a binary feature `unusual_tcp_flags`
top_3_tcp_flags = df_engineered['tcp_flags'].value_counts().nlargest(3).index.tolist()
df_engineered['unusual_tcp_flags'] = df_engineered['tcp_flags'].apply(lambda x: 0 if x in top_3_tcp_flags else 1)

# 7. Create a binary feature `mqtt_conflags_anomaly`
top_3_mqtt_conflags = df_engineered['mqtt_conflags'].value_counts().nlargest(3).index.tolist()
df_engineered['mqtt_conflags_anomaly'] = df_engineered['mqtt_conflags'].apply(lambda x: 0 if x in top_3_mqtt_conflags else 1)

# 8. Create a binary feature `mqtt_kalive_deviation`
df_engineered['mqtt_kalive_deviation'] = df_engineered['mqtt_kalive'].apply(lambda x: 1 if (x != 0 and x != 60) else 0)

# 9. Create a binary feature `high_mqtt_qos`
df_engineered['high_mqtt_qos'] = df_engineered['mqtt_qos'].apply(lambda x: 1 if x == 2 else 0)

# 10. Convert the `timestamp` column in `df_engineered` to datetime objects and sort
df_engineered['timestamp'] = pd.to_datetime(df_engineered['timestamp'])
df_engineered = df_engineered.sort_values(by='timestamp').reset_index(drop=True)

# 11. Calculate the `inter_arrival_time` feature
df_engineered['inter_arrival_time'] = df_engineered['timestamp'].diff().dt.total_seconds().fillna(0)

# 12. Extract the `time_of_day_seconds` feature
df_engineered['time_of_day_seconds'] = df_engineered['timestamp'].dt.hour * 3600 + \
                                     df_engineered['timestamp'].dt.minute * 60 + \
                                     df_engineered['timestamp'].dt.second

# 13. Extract the `day_of_week` feature
df_engineered['day_of_week'] = df_engineered['timestamp'].dt.dayofweek

# 14. Remove the `mqtt_dupflag` column
df_engineered = df_engineered.drop(columns=['mqtt_dupflag'])

# 15. Display the first 5 rows of `df_engineered`
print("Engineered DataFrame Head:")
display(df_engineered.head())

# 16. Print a concise summary of `df_engineered`
print("\nEngineered DataFrame Info:")
df_engineered.info()

# 17. Print the value counts for the new categorical features
print("\nValue Counts for New Categorical/Binary Features:")
print("\ntcp_len_category:")
print(df_engineered['tcp_len_category'].value_counts())
print("\nmqtt_msg_category:")
print(df_engineered['mqtt_msg_category'].value_counts())
print("\nunusual_tcp_flags:")
print(df_engineered['unusual_tcp_flags'].value_counts())
print("\nmqtt_conflags_anomaly:")
print(df_engineered['mqtt_conflags_anomaly'].value_counts())
print("\nmqtt_kalive_deviation:")
print(df_engineered['mqtt_kalive_deviation'].value_counts())
print("\nhigh_mqtt_qos:")
print(df_engineered['high_mqtt_qos'].value_counts())
````
